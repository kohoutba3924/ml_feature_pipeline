```python
# Imports and file path
import duckdb
from pathlib import Path

parquet_path = Path("../data/external/lcdv2_stations.parquet")
parquet_path
```

```python
# Preview first 10 rows
duckdb.query(f"""
    SELECT *
    FROM '{parquet_path.as_posix()}'
    LIMIT 10
""").df()
```

```python
# Count stations
duckdb.query(f"""
    SELECT COUNT(*) AS station_count
    FROM '{parquet_path.as_posix()}'
""").df()
```

```python
# Highest elevation stations
duckdb.query(f"""
    SELECT station, name, elevation
    FROM '{parquet_path.as_posix()}'
    ORDER BY elevation DESC
    LIMIT 10
""").df()
```

```python
# Stations by decade of first observation
duckdb.query(f"""
    SELECT
        CAST(substr(min_date, 1, 4) AS INTEGER) / 10 * 10 AS decade,
        COUNT(*) AS station_count
    FROM '{parquet_path.as_posix()}'
    GROUP BY decade
    ORDER BY decade
""").df()
```

```python
# Convert DuckDB query to Pandas DataFrame
df_duck = duckdb.query(f"SELECT * FROM '{parquet_path.as_posix()}'").df()
df_duck.head()
```
