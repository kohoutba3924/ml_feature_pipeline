```python
# Imports and file path
import duckdb
from pathlib import Path

acs5_parquet_path = Path("../data/normalized/acs5_2024.parquet")
tiger_parquet_path = Path("../data/normalized/tiger_geospatial_tracts.parquet")
lcdv2_parquet_path = Path("../data/normalized/lcdv2/station_id=WBAN_00183/year=2025/part.parquet")
station_elevation = Path("../data/raw/elevation/station_elevation.parquet")
tract_elevation = Path("../data/raw/elevation/tract_elevation.parquet")
```

```python
# Preview first 10 rows
duckdb.query(f"""
    SELECT *
    FROM '{acs5_parquet_path.as_posix()}'
    LIMIT 10
""").df()
```

```python
# Record Count
duckdb.query(f"""
    SELECT COUNT(*) AS station_count
    FROM '{acs5_parquet_path.as_posix()}'
""").df()
```

```python
# Most recently available dates from most recently available year
duckdb.query(f"""
    SELECT station, timestamp
    FROM '{lcdv2_parquet_path.as_posix()}'
    ORDER BY timestamp DESC
    LIMIT 50
""").df()
```
