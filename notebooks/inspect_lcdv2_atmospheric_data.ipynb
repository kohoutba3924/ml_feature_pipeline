```python
import polars as pl
from pathlib import Path

# Load the CSV (all columns as Utf8 to avoid inference issues)

csv_path = Path("../data/raw/lcdv2/hourly/2024/72034354852-2024.csv")

df = pl.read_csv(
    csv_path,
    infer_schema_length=0,   # force all Utf8
    schema_overrides={"DATE": pl.Utf8, "STATION": pl.Utf8}
)

df.head()
```

```python
cols = df.columns

daily_fields = [c for c in df.columns if c.startswith("Daily")]
daily_fields

```

```python
# Convert DATE to a datetime type & some date component fields for grouping/sorting
df = (
    df
    .with_columns([
        pl.col("DATE").str.strptime(pl.Datetime, strict=False).alias("dt"),
    ])
    .with_columns([
        pl.col("dt").dt.date().alias("date"),
        pl.col("dt").dt.month().alias("month"),
    ])
)

df.select(["DATE", "dt", "date", "month"]).head()
```

```python
# Boolean: does this row have ANY populated daily field?
df = df.with_columns(
    pl.any_horizontal([pl.col(f) != "" for f in daily_fields]).alias("has_daily_values")
)

df.select(["DATE", "has_daily_values"] + daily_fields).head()

```

```python
# Count rows per day that contain daily field values
daily_counts = (
    df.group_by("date")
      .agg(pl.col("has_daily_values").sum().alias("daily_rows_with_values"))
      .sort("date")
)

daily_counts

```

```python
# Days with no daily value entries
days_with_zero = daily_counts.filter(pl.col("daily_rows_with_values") == 0)
days_with_zero
```

```python
# Days with rows containing daily values > 1
days_with_multiple = daily_counts.filter(pl.col("daily_rows_with_values") > 1)
days_with_multiple
```

Appears as though daily values are only entered once per day at maximum, but are not reliably entered once per day.

Will check the consistency and quality of hourly fields.

```python
# Check for duplicate timestamps
dupes = (
    df.group_by("dt")
      .agg(pl.len().alias("count"))
      .filter(pl.col("count") > 1)
)

dupes
```

```python
# Checking for time interval consistency
df_sorted = df.sort("dt")

df_gaps = (
    df_sorted
    .with_columns(
        pl.col("dt").diff().alias("delta")
    )
    .select(["DATE", "dt", "delta"])
)

df_gaps.head(20)
```

```python
# Looks consistent at a glance, checking for large gaps
large_gaps = df_gaps.filter(pl.col("delta") > pl.duration(hours=2))
large_gaps
```

```python
summary = {
    "days_with_zero_daily_rows": days_with_zero.height,
    "days_with_multiple_daily_rows": days_with_multiple.height,
    "duplicate_timestamps": dupes.height,
    "large_gaps": large_gaps.height,
}

summary
```

Daily data is unacceptabel, hourly data is seemingly acceptable. Repeating with one more station sample.
